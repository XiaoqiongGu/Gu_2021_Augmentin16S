## Figure 1 plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
metadata = pd.read_csv('data/metadata_BS.csv',index_col=0)
metadata = metadata.set_index('S' + metadata.index.astype(str)).reset_index()
metadata.head()
```

```python
# Figure 1B - scatter plot
plt.style.use('seaborn-white')
fig,ax=plt.subplots(figsize=(10,10))
ax = sns.scatterplot(data=metadata, x='total_number_of_episodes_d1_d2_d3', y='maximum_BS_d1_d2_d3',
                     hue='Group', size='Freq', sizes=(400, 2000), palette=['#80b1d3', '#fb8072'])
plt.xticks(np.arange(0, 20, step=4),fontsize= 26)
plt.yticks(np.arange(1,8),fontsize= 26)

h,l = ax.get_legend_handles_labels()
plt.legend(h[1:3],l[1:3],loc='lower right', borderaxespad=0.,fontsize=34, markerscale=3) #only choose AAD and non-AAD as labels

ax.tick_params(axis='both',labelsize=34,direction="in")
ax.set_xlabel('Total number of episodes',size=34) #across day 1, 2 and 3
ax.set_ylabel('Maximum of Bristol Stool Scale',size=34) #across day 1, 2 and 3

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(4)  

ax.grid(axis='both', which='major', ls='--')
plt.savefig('plots/Figure1B_max_BS_vs_episodes.svg',dpi=300)
```
